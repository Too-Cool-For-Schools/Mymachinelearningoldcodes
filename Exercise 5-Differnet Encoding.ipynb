Here is the key summary of differnet encoding types:

Code below show the **ordinal-encoding** where ranking of categories matter with more than two ranks. 
    </div>

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data=fetch_openml('car',as_frame=True).frame
```

```python
data
```

```python
columns_to_encode=['lug_boot','safety'] #define the column names first
encoder=OrdinalEncoder(categories=[
    ['small','med','big'],
    ['low','med','high'],
                                  ])   #define the categories values inside those columns

data[columns_to_encode]=encoder.fit_transform(data[columns_to_encode]) #convert them to numerical encoding using encoder.fit_transform
```

```python
data
```

```python
encoder.inverse_transform(data[columns_to_encode])
```
